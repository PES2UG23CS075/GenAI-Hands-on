##Assignment 2: Recursive Function Generator (Prompt Engineering)
**Objective**

Use the CO-STAR framework and Structured Prompting to force the LLM to act as a Senior Developer and write a specific, constrained Python function.

Constraints
Persona: Senior Python Developer.

Logic: Must use Recursion.

Restriction: Absolutely no string slicing ([::-1]) for the reversal logic.

Format: Output only the code block with professional documentation.


```python
%pip install -U langchain-google-genai --quiet

import getpass
import os
from langchain_google_genai import ChatGoogleGenerativeAI

if "GOOGLE_API_KEY" not in os.environ:
    os.environ["GOOGLE_API_KEY"] = getpass.getpass("Enter your Google API Key: ")

# Initialize the model with low temperature for coding tasks (precision over creativity)
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash", temperature=0.0)
```

```python
# THE ASSIGNMENT: Structured Prompt for a Python Function
structured_prompt = """
# Context
You are a Senior Python Developer with expertise in writing clean,
maintainable, and well-documented code.

# Objective
Write a Python function that reverses a given string.

# Constraints
1. The function MUST use recursion.
2. Do NOT use string slicing (e.g., [::-1]).
3. Do NOT use any built-in reverse functions.

# Style Guidelines
Include a detailed docstring explaining the purpose, parameters, return value, and an example.

# Output Format
Return ONLY the Python code block. No conversational filler.
"""

# Run the prompt
response = llm.invoke(structured_prompt)
print(response.content)
```
